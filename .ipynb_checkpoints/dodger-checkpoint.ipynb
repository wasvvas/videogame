```python
import pygame, random, sys
from pygame.locals import *
pygame.init()

#screen
res = (600, 600)
screen = pygame.display.set_mode(res)
pygame.display.set_caption('dodger')

#constant variables
txtcolor = (240,240,255)
fps = 20
rockminsize = 30
rockmaxsize = 70
rockminspeed = 3
rockmaxspeed = 10
addnewrock = 10
playerspeed = 9

#functions
def playerhitrock(playerRect, rocks): 
    for r in rocks:
        if playerRect.colliderect(r['rect']):
            return True
    return False

def pressspacetostart():
    start = True
    while start:
        for event in pygame.event.get():
            if event.type == KEYDOWN:
                if event.key == K_SPACE:
                    start = False

def drawtext(text, font, surface, x, y):
    textobj = font.render(text, 1, txtcolor)
    textrect = textobj.get_rect()
    textrect.center = (x, y)
    surface.blit(textobj, textrect)

def quitgame():
    pygame.display.quit()
    pygame.quit()
    sys.exit()

def maingame():
    #game setup
    font = pygame.font.SysFont(None, 40)
    mainClock = pygame.time.Clock() # in case it goes too fast
    player = pygame.image.load('player-01.png')
    playerRect = player.get_rect()
    background = pygame.image.load('background-01.png')
    bgRect = background.get_rect()
    start = pygame.image.load('start-01.png')
    startRect = start.get_rect()
    rock = pygame.image.load('rock.png')
    pygame.display.update()
    rocks=[]
    score=0
    scorecounter=0
    playerRect.topleft=(250, 550)
    moveL=moveR=moveU=moveD=False
    addrockcounter=0

    #start screen
    screen.blit(start,startRect)
    pygame.display.update()
    pressspacetostart()

    #game loop
    while True:
    
        #movement keys
        for event in pygame.event.get():
            if event.type==KEYDOWN:
                if event.key==ord('a'):
                    moveL=True
                    moveR=False #sets opposite direction to false to prevent freezing
                if event.key==ord('d'):
                    moveR=True
                    moveL=False
                if event.key==ord('w'):
                    moveU=True
                    moveD=False
                if event.key==ord('s'):
                    moveD=True
                    moveU=False
            if event.type==KEYUP:
                if event.key==ord('a'):
                    moveL=False 
                if event.key==ord('d'):
                    moveR=False
                if event.key==ord('w'):
                    moveU=False
                if event.key==ord('s'):
                    moveD=False
        #scoring
        scorecounter+=1
        if scorecounter == fps: # 1 point per second
            score+=1
            scorecounter=0

        #generating new rocks
        addrockcounter+=1
        if addrockcounter == addnewrock:
            addrockcounter=0
            rocksize = random.randint(rockminsize,rockmaxsize)
            newrock = {'rect': pygame.Rect(random.randint(0, 600-rocksize), 0, rocksize, rocksize),
                      'speed':random.randint(rockminspeed, rockmaxspeed),
                      'surface':pygame.transform.scale(rock, (rocksize, rocksize))}
            rocks.append(newrock)
    
        #moving player
        if moveL and playerRect.left > 0:
            playerRect.move_ip(-1 * playerspeed, 0)
        if moveR and playerRect.right < 600:
            playerRect.move_ip(playerspeed, 0)
        if moveU and playerRect.top > 0:
            playerRect.move_ip(0, -1 * playerspeed)
        if moveD and playerRect.bottom < 600:
            playerRect.move_ip(0, playerspeed)

        #moving rocks
        for r in rocks:
            r['rect'].move_ip(0, r['speed'])

        #removing rocks after they fall
        for r in rocks[:]:
            if r['rect'].top > 600:
                rocks.remove(r)

        #update screen
        screen.blit(background, bgRect)
        drawtext('Score: %s' % (score), font, screen, 75, 25)
        screen.blit(player, playerRect)
        for r in rocks:
            screen.blit(r['surface'], r['rect'])
        pygame.display.update()

        #game over
        if playerhitrock(playerRect, rocks):
            break

        mainClock.tick(fps)
    #end screen - quit or restart
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == QUIT:
                quitgame()
                running = False
            else:
                pressspacetostart()
                maingame()
                running = False
maingame()
```
